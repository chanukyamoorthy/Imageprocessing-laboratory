```python
#Discrete function Transformation
import numpy as np
import matplotlib.pyplot as plt
mat=[]

for k in range(8):
    temp = []
    for n in range(8):
        temp.append(np.exp(-1j*k*n/8))
    mat.append(temp) 
mat[7]
```

```python
mat=np.asarray(mat)
trmat=[[0 for i in range(8)] for j in range(8)]
for i in range (8):
    for j in range (8):
        trmat[i][j]=np.matmul(mat[i].reshape(8,1),mat[j].reshape(1,8))

trmat=np.array(trmat)
```

```python
realpart=np.zeros(trmat.shape)
imgpart=np.zeros(trmat.shape)
for i in range(8):
    for j in range(8):
        realpart[i,j]=trmat[i,j].real
        imgpart[i,j]=trmat[i,j].imag
```

```python
# plt.figure(figsize=(8,8))
for i in range(8):
    for j in range(8):
        plt.figure()
        plt.imshow(realpart[i][j],cmap='gray')
        
```
