# CONSTANT STRETCHING

```python
img3=cv2.imread('C:/Users/STUDENT.SW3-C21/Desktop/116cs0192/grain.png',0)
plt.imshow(img3,cmap='gray')
```

```python
xa=80
xb=160
xc=255
ya=40
yb=200
yc=255

x=40/80
y=160/80
z=55/(255-160)
n=img3.shape[0]
m=img3.shape[1]
newimg=np.zeros((m,n))
for i in range(0,m):
    for j in range(0,n):
        if(img3[i,j]<xa):
            newimg[i,j]=0.5*img3[i,j]
        elif(img3[i,j]>xa and img3[i,j]<xb):
            newimg[i,j]=2*(img3[i,j]-80)+40
        elif(img3[i,j]>xb and img3[i,j]<xc):
            newimg[i,j]=0.578*(img3[i,j]-160)+200
            
```

```python
n=img3.shape[0]
m=img3.shape[1]
ar=np.zeros((256),np.uint8)
for i in range (0,n):
    for j in range(0,m):
        val=img3[i,j]
        ar[val]+=1

plt.bar(np.arange(len(ar)),ar)
```

```python
n=newimg.shape[0]
m=newimg.shape[1]
ar1=np.zeros((256),np.uint8)
for i in range (0,n):
    for j in range(0,m):
        val=newimg[i,j]
        ar1[int(val)]+=1

plt.bar(np.arange(len(ar1)),ar1)
```
