# HUFFMAN CODING

```python
class Tree:
    def __init__(self):
        self.freq = None
        self.char = None
        self.left = None
        self.right = None
        
    def set_data(self, data):
        self.char = data
        
    def set_freq(self, freq):
        self.freq = freq
        
    def __lt__(self, other):
        return self.freq < other.freq
    
    def __str__(self):
        return str(self.char) + ' ' + str(self.freq)
```

```python
from heapq import heappop, heappush
import numpy as np
```

```python
def dfs(node, string):
    if node.left is None and node.right is None:
        code[node.char] = string
        return None
    
#     print(node)
    if node.left is not None:
        dfs(node.left, string + '0')
    if node.right is not None:
        dfs(node.right, string + '1')
```

```python
def encoding(data):
    
    freq = {}
    data = data.flatten()
    for ele in data:
        if ele in freq.keys():
            freq[ele] += 1
        else:
            freq[ele] = 1
            
    print(freq)
            
    leaves = []
    chars = []
    for key, value in freq.items():
        chars.append((key, value))
        
    for char in chars:
        temp = Tree()
        temp.set_data(char[0])
        temp.set_freq(char[1])
        leaves.append(temp)
        
    heap = []
    
    for leaf in leaves:
        heappush(heap, leaf)
        
    while len(heap) > 1:
        ele1, ele2 = heappop(heap), heappop(heap)
        temp = Tree()
        temp.freq = ele1.freq + ele2.freq
        temp.left = ele1
        temp.right = ele2

        heappush(heap, temp)
        
    
    dfs(heap[0], '')
    
    return freq
```

## Arithmetic Coding

```python
ranges=[(0,0.4),(0.4,0.6),(0.6,0.7),(0.7,1)]
final_ranges=[ranges]
```

```python
string = 'aad'
```

```python
for char in string:
    ele = ord(char) - ord('a')
    low = ranges[ele][0]
    
    temp = []
    diff = ranges[ele][1] - ranges[ele][0]
    
    for i in range(len(ranges)):
        low_temp = low + diff * final_ranges[-1][i][0]
        high_temp = low + diff * final_ranges[-1][i][1]

        temp.append((low_temp, high_temp))
        
    ranges = temp
    final_ranges.append(temp)
```
