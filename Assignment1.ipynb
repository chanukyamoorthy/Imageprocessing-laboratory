```python
import numpy as np
import cv2
import math
import matplotlib.pyplot as plt
img = np.zeros((512,512),np.uint8)
d=20
for i in range (50+d,350+d):
    for j in range (50+d,350+d):
        img[i][j]=1;
for i in range (70+d,330+d):
    for j in range (70+d,330+d):
        img[i][j]=0;
        
```

```python
  
plt.imshow(img,cmap='gray')
```

```python
import numpy as np
import cv2
import math
import matplotlib.pyplot as plt
img = np.zeros((512,512),np.uint8)
    
for x in range(0,511):
    for y in range(0,511):
        img[x][y]=255*math.cos(2*(x/50 + y/25));
  
        

    
plt.imshow(img,cmap='gray')      
```

```python
img = np.zeros((512,512),np.uint8)
r=150
x=256
y=256
for i in range(0,511):
    for j in range(0,511):
        if ((x-i)*(x-i)+(y-j)*(y-j)) < r*r:
            img[i][j]=1;
d=130       
for i in range(0,511):
    for j in range(0,511):
        if ((x-i)*(x-i)+(y-j)*(y-j)) < d*d:
            img[i][j]=0;            
  
        

    
plt.imshow(img,cmap='gray')  
```
